2. K-MEAN

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans()# default 8 Cluster
yp = km.fit_predict(x)
yp
```

```python
len(yp)
```

```python
km.cluster_centers_
```

```python
# Elbow curve 

inertia = []
K= range(1,11)

for k in K:
    kmean= KMeans(n_clusters= k, random_state=42)
    kmean.fit(x)
    inertia.append(kmean.inertia_)
plt.plot(K,inertia)
plt.xlabel('number of cluster')
plt.ylabel('inertia')
plt.title('elbow method for optimal k')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x,ypred)
```

```python
scores = []
K = range(2, 11)  

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(x)
    score = silhouette_score(x, labels)
    scores.append(score)
    print(f"For k={k}, silhouette score={score:.3f}")
plt.plot(K, scores, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
Kmean= KMeans(n_clusters = 2, random_state= 42)
label= Kmean.fit_predict(x)

df['cluster']= label
df.head()
```

# Conclusion:



The purpose of this assignment was to learn and apply three popular clustering algorithms — K-Means, Hierarchical Clustering, and DBSCAN = to group similar data points without labeled outcomes. We first conducted preprocessing and exploratory data analysis (EDA) to better understand the dataset before clustering.

During the EDA process:

The dataset was clean with no missing entries.

There were no significant outliers detected in the initial assessment.

Feature scaling was applied to ensure balanced clustering results.

Distribution patterns were explored through histograms and scatter plots.

Relationships between variables were examined using a correlation heatmap.

The dataset showed promising patterns that could be grouped into clusters.

For each clustering method:
K-Means Clustering:
The Elbow Method helped in selecting the optimal number of clusters, where k = 2 appeared to be the most appropriate. The silhouette score at this k was 0.325, indicating moderate cluster separation.

Hierarchical Clustering:
Different linkage methods such as single, complete, average, and ward were tested. The dendrogram helped visualize how the data grouped together, with the ward linkage method providing the clearest separation.

DBSCAN:
Parameters such as eps = 0.5 and min_samples = 5 were used. This method successfully identified clusters along with noise points that didn’t belong to any cluster.

Comparison of Methods:

K-Means provided effective clustering but required the number of clusters to be defined in advance.

Hierarchical clustering offered a comprehensive view of cluster relationships but may not be efficient with larger datasets.

DBSCAN was useful in detecting outliers but needed careful tuning of parameters for best performance.

Overall, each algorithm showed its strengths and limitations, and applying multiple clustering methods allowed a thorough understanding of the dataset’s structure.
